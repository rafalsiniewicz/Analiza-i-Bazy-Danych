```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px

```

```python
weather=pd.read_csv('weather_tidy.csv',delimiter=',')
print(weather.to_string())
```

```python
######## najwieksza roznica temp ###########
# ze wzgledu na fakt, ze jest tylko jedna stacja pogodowa jako wynik podaje wszystkie informacje
# a w szczegolnosci date wystapienia danej temp, a nie tylko id stacji pogodowej
# gdyz nie jest to jednoznaczna informacja
print()
print("najwieksza roznica temp")
weather['diff'] = weather['tmax'] - weather['tmin']
print(weather.loc[weather['diff'].idxmax()])
```

```python
######## najblizsza tmax do sredniej tmax ##########################################
print()
print("najblizsza tmax do sredniej tmax")
weather = weather.dropna()
print(weather.iloc[(weather['tmax'] - weather['tmax'].mean()).abs().argsort()[:1]])
print("srednia tmax ", weather['tmax'].mean())
```

```python
######## najblizsza tmin do sredniej tmin ##########################################
print()
print("najblizsza tmin do sredniej tmin")
print(weather.iloc[(weather['tmin'] - weather['tmin'].mean()).abs().argsort()[:1]])
print("srednia tmin", weather['tmin'].mean())
```
